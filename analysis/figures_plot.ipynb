```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_train_test_class_dist(file_name):
    train_class_dist = {}
    file = open(file_name)
    file.readline()
    line = file.readline()
    while line.startswith('CLASS'):
        dat = line.split(' ')
        train_class_dist[dat[1]] = dat[4]

        line=file.readline()

    test_class_dist = {}
    while line.startswith('CLASS'):
        dat = line.split(' ')
        test_class_dist[dat[1]] = dat[4]

        line=file.readline()
    
    return train_class_dist,test_class_dist

def read_true_predicted_counts(file_name):
    train_class_dist = {}
    file = open(file_name)
    file.readline()
    line = file.readline()
    while line.startswith('Class'):
        dat = line.split(' ')
        train_class_dist[dat[1]] = dat[5][:-1]

        line=file.readline()

    test_class_dist = {}
    line = file.readline()
    while line.startswith('Class'):
        dat = line.split(' ')
        test_class_dist[dat[1]] = dat[5][:-1]

        line=file.readline()
    
    return train_class_dist,test_class_dist
```
